```python
df_add_Mrt_cbd_inf = df_addressCoorExtCSV.copy()
```

```python
#creating list for latitude and longitude for Property address and MRTS 
lst_add_lat = df_add_Mrt_cbd_inf['latitude']
lst_add_long = df_add_Mrt_cbd_inf['longitude']
lst_mrt_lat = df_mrt_stations['latitude']
lst_mrt_long = df_mrt_stations['longitude']
```

```python
# Combining the Latitude and Longitude  List together 
PropAdd_coOrd_lst = []
MRTS_coOrd_lst = []

for lat, long in zip(lst_add_lat, lst_add_long):
    PropAdd_coOrd_lst.append((lat,long))
for lat, long in zip(lst_mrt_lat, lst_mrt_long):
    MRTS_coOrd_lst.append((lat, long))
```

# Getting the proximity Values with geopy geodesic

```python
# Distance to nearest MRT
from geopy.distance import geodesic

list_of_dist_mrt = []
min_dist_mrt = []

for origin in PropAdd_coOrd_lst:
    for destination in range(0, len(MRTS_coOrd_lst)):
        list_of_dist_mrt.append(geodesic(origin,MRTS_coOrd_lst[destination]).meters)
    shortest = (min(list_of_dist_mrt))
    min_dist_mrt.append(shortest)
    list_of_dist_mrt.clear()
```

```python
# Distance from CDB
cbd_dist = []

for origin in PropAdd_coOrd_lst:
    cbd_dist.append(geodesic(origin,(1.2830, 103.8513)).meters) #CBD coordinates
```

```python
# Put MRT and CBD distance together
df_add_Mrt_cbd_inf['cbd_Dist'] = cbd_dist
df_add_Mrt_cbd_inf['mrt_minDist'] = min_dist_mrt
```
